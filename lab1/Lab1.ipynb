```python
from matplotlib import pyplot as plt
import numpy as np
```

Polecenie:

Narysuj wykresy funkcji $f(x)=x^2+5$ dla poniższych przedziałów:

1) $x>-1$ && $x<1$

```python
x = np.linspace(-1,1,100)
f = x**2 + 5
plt.plot(x,f)
plt.xlabel("Arguments")
plt.ylabel("Values")
plt.title("$f(x)=x^2+5$ for x=(-1,1)")
plt.legend(["f(x)"])
```

2) $x>-6$ i $x<6$

```python
x = np.linspace(-6,6,200)
f = x**2 + 5
plt.plot(x,f)
plt.xlabel("Arguments")
plt.ylabel("Values")
plt.title("$f(x)=x^2+5$ for x=(-6,6)")
plt.legend(["f(x)"])
```

3) $x>0$ i $x<5$

```python
x = np.linspace(0,5,200)
f = x**2 + 5
plt.plot(x,f)
plt.xlabel("Arguments")
plt.ylabel("Values")
plt.title("$f(x)=x^2+5$ for x=(0,5)")
plt.legend(["f(x)"])
```
